```python
# test_10000.csv


import numpy as np
import pandas as pd
import datetime as dt
# import matplotlib.pyplot as plt
#[]
cols = ['JRNY_TYPE_ID','ORIG','DEST','COMPARTMENT','PAX_TYPE','SALE_STR_DATE','TRAVEL_STR_DATE','SEASON_STR_DATE','DOW','MIN_STAY_UNIT','MIN_STAY_VAL','MAX_STAY_UNIT',
        'MAX_STAY_VAL','AP_UNIT','AP_VALUE','RBD']
res_col = ['']
path = 'test_10000.csv'
df = pd.read_csv(path, usecols=cols)


```

```python

def year_to_int(date):
    return date.timetuple().tm_yday
    
#test it
# today = dt.datetime.now()
# print(today)
# print(year_to_int(today))
```

```python
def convert_rev_date(cur):
        if cur is int : 
            print("already converted")
            return cur
#         print("test")
#         print(cur, "-_-")
#         print("will splt")
        [day,month,year] = [int(x) for x in cur.split('-')]
#         print("split")
        date = dt.datetime(day=day%100,month=month, year=year+2000)
#         print("made date")
        date = year_to_int(date)
#         print("made day/year")
#         print(date)
        return date 

def convert_date_for(features):
    for feature in features:
        for i in range(df[feature].size):
            cur = df.loc[i,feature]
            df.loc[i, feature] = convert_rev_date(cur)
#             [day,month,year] = [int(x) for x in cur.split('-')]
#             date = dt.datetime(day=day-2000,month=month, year=year+2000)
#             date = year_to_int(date)
#     #         df[col][i] = date
#             data.loc[:, each] = (data[each] - mean)/std
#             df.loc[:,(col,i)] = date
            
#         if i > 10: next
#         print(date)
#         print(df[col][i])
```
